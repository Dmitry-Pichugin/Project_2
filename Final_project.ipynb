### Функция для первичного анализа данных.

Создадим функцию для одновременного вывода основного описания датасета

```python
def desc(df):

    print('\033[1m  Общая информация:  \033[0m')
    display(df.info())
    print('------------------------------------------------------------------')
    
    print('\033[1m  Статистическое описание столбцов:  \033[0m')
    display(df.describe(include='all'))
    print('------------------------------------------------------------------')
    
    print('\033[1m  Пропуски в столбцах:  \033[0m')
    nan_count = df.isna().sum().reset_index()
    nan_count = nan_count.rename({'index': 'columns', 0: 'nan_count'}, axis=1)
    nan_count['nan_share'] = round(nan_count['nan_count'] * 100 / nan_count['nan_count'].sum(), 2)
    display(nan_count)
    print('------------------------------------------------------------------')
    
    print('\033[1m  Уникальные значения в столбцах:  \033[0m')    
    values = pd.DataFrame()
    for column in df.columns:
        values[column] = [df[column].unique(), df[column].nunique()]
    display(values.T.rename({0: 'list_of_values', 1: 'unique_values_number'}, axis=1))
    print('------------------------------------------------------------------')
    
    print('\033[1m  Явных дублкатов обнаружено:  \033[0m' + str(df.duplicated().sum()))
```

Добавлю данные со статусом и видом завершения первого уровня в данные с игроками.

```python
users.loc[users['user_id'].isin(passed_pvp), 'st_1_status'] = 'passed_pvp'
users.loc[users['user_id'].isin(passed_pve), 'st_1_status'] = 'passed_pve'

users.head()
```

```python
users.st_1_status.value_counts()
```

Все правильно

Добавляем столбцы:
- Время первой и последней активности на первом уровне.
- Время, проверенное игроком на первой стадии (разницы последней и первой активностью) в числовом формате.
- Количество дней, проведенных на первом уровне с округлением в большую сторону.
- Количество активных дней.
- Количество построек у каждого игрока.

Кроме того, поскольку активные пользователи переходят на следующий уровень за разное время, а нас интересует количество построек у игроков, как потенциально единственный источник рекламы, посчитаем среднее количество построек в день активности.  
Для игроков совершивших только одну постройку укажем количество построек в день равное 0.

```python
users[['first_active', 'last_active', 'active_days', 'building_count']] = users.merge(
    logs.groupby('user_id').agg({'event_datetime': {'max', 'min'}, 'date': 'nunique', 'building_type': 'count'}),
    left_on='user_id', right_index=True
).iloc[:, -4:]

users['first_st_dur_days'] = (users['last_active'] - users['first_active'])  / timedelta(days=1)
users['days_played'] = np.ceil(users['first_st_dur_days'])

users['building_per_day'] = users['building_count'] / users['days_played']
users.loc[users['building_count'] == 1, 'building_per_day'] = 0

users.head()
```

Проверим, все ли данные подтянулись и формат данных

```python
users.info()
```

Все подтянулось корректно.

Добавим в колонку со статусом прохождения первой стадии игроками данные о текущей активности или выбытии игрока.  
Для этого мы установим трешхолд активности игроков за 7 дней до последнего известного лога. Если клиент не был активен последнюю неделю, его можно считать выбывшим.

```python
threshold = users.last_active.max() - timedelta(days=7)
threshold
```

Теперь добавим данные с активностью игроков.

```python
users.loc[users['st_1_status'].isnull() & (users['last_active'] < threshold), 'st_1_status'] = 'inactive'
users.loc[users['st_1_status'].isnull() & (users['last_active'] >= threshold), 'st_1_status'] = 'still_active'

users.info()
```

Все необходимые данные заполнены корректно.

### Создание функций

Для упрощения дальнейшей работы, создадим функции для формирования сводных таблиц и их более удобного вывода.

Создадим функцию для подсчета сводных таблиц с процентами

```python
def to_pivot(df, index, values='user_id', aggfunc='count'):
    pivot = df.pivot_table(index=index, values=values, aggfunc=aggfunc, margins=True)
    pivot = pivot.rename(columns={values: aggfunc})
    pivot['share'] = round(pivot * 100 / pivot.loc['All',aggfunc], 2)
    pivot = pivot.sort_values(by=aggfunc, ascending=False)
    return pivot
```

Создадим функцию для форматирования таблиц.

```python
def format_pivot(df):
    return(df.style.format({i: '{:.2f}' for i in df.columns}) \
        .highlight_max(color='yellowgreen', subset=df.columns) \
        .highlight_min(color='coral', subset=df.columns)
          )
```

## Проверка гипотез

### Гипотеза: Время завершения уровня различается в зависимости способа прохождения

Нулевая гипотеза: Время завершения уровня не различается в зависимости способа прохождения.  

Альтернативная гипотеза: Время завершения уровня различается в зависимости способа прохождения

Изучим статистические данные по типу завершения первого уровня.

```python
round(users.pivot_table(index='st_1_status', values='first_st_dur_days', aggfunc = (['count', 'var']))[1:3], 2)
```

```python
alpha = .05

results = st.ttest_ind(users[users.st_1_status == 'passed_pvp']['first_st_dur_days'],
                       users[users.st_1_status == 'passed_pve']['first_st_dur_days'], equal_var=False)[1]

results
```

```python
if results < alpha:
    display("Отвергаем нулевую гипотезу")
else:
    display("Не получилось отвергнуть нулевую гипотезу") 
```

P-value значительно ниже критического уровеня статистической значимости из-за чего мы можем с уверенностью отвергнуть нулевую гипотезу и признать, что время прохождения уровня у PvP и PvE игроков разное.

### Гипотеза: Количество построек у игроков отличается в зависимости от канала привлечения.

- Нулевая гипотеза: Количество построек зданий одинаково для игроков, перешедших из разных каналов.
- Альтернативная гипотеза: Количество построек зданий отличается в зависимости от канала привлечения.
- Альфу примем 5%.

```python
round(users.pivot_table(index='source', values='building_count', aggfunc = (['count', 'var']), margins=True), 2)
```

Посчитаем разницу между максимальным и минимальным значениями.

```python
round(users.pivot_table(index='source', values='building_count', aggfunc = (['count', 'var'])).max() / (
users.pivot_table(index='source', values='building_count', aggfunc = (['count', 'var'])).min()) - 1, 2)
```

Проверяем распределение выборок на нормальность критерием Шапиро-Уилка.

- Нулевая гипотеза: Данные нормально распределены.
- Альтернативная гипотеза: данные распределены не нормально.
- Альфу примем 1%.

```python
alfa = 0.01
values = pd.DataFrame()

for source in users.source.unique():
    results = st.shapiro(users[users.source == source]['building_count'])
    values[source] = results
    
values = values.T
values['result'] = values[1].apply(lambda x: 'Отвергаем нулевую гипотезу' if x < alpha else 'Не отвергаем')

values = values.iloc[:, 1:].rename(columns={1: 'p-value'})

values
```

- Данные распределены не нормально.
- Разница количества пользователей в группах доходит до 80%.
- Разница дисперсии не превышает 3%

В связи с этим, для проверки гипотезы о равенстве времени на прохождение уровня мы используем тест Манна-Уитни.  
Альфу примем 5%

```python
p_table = pd.DataFrame()

for source, i in zip(users['source'].unique()[:-1], range(3)):
    values = pd.DataFrame()
    for source_1 in users['source'].unique()[1:]:
        results = st.mannwhitneyu(users[users['source'] == source_1]['building_count'],
                                  users[users['source'] == source]['building_count'])
        values[source_1] = results
    p_table[source] = values.T[1][i:]

round(p_table, 3)
```

```python
alpha = .05

p_table_result = pd.DataFrame()

for column, i in zip(p_table.columns, range(3)):
    p_table_result[column] = p_table[column].map(lambda x: 'Отвергаем нулевую гипотезу' if x < alpha else 'Не отвергаем')[i:]
    
p_table_result
```

Исходя из текущих данных, можно отвергнуть гипотезу о равенстве количества построек у игроков, перешедших из разных каналов. 
Поскольку мы производим несколько сравнений, применим поправку Шидака для изменения принятого уровня значимости Альфа.

```python
experiments_count = p_table_result.count().sum()
experiments_count
```

```python
alpha = 1 - (1 - alpha)**(1/experiments_count)
alpha
```

```python
p_table_result = pd.DataFrame()

for column, i in zip(p_table.columns, range(3)):
    p_table_result[column] = p_table[column].map(lambda x: 'Отвергаем нулевую гипотезу' if x < alpha else 'Не отвергаем')[i:]
    
p_table_result
```

**Вывод:** С учетом поправки Шидика мы не может отвергнуть гипотезу о равенстве количества построек у игроков из разных каналов привлечения.
